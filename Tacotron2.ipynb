```python
import os
from tqdm import tqdm
from datasets import preprocessor
from hparams import hparams

def generate_train_txt(input_folders, output_folder, n_jobs):
    mel_dir = os.path.join(output_folder, 'mels')
    linear_dir = os.path.join(output_folder, 'linear')
    wav_dir = os.path.join(output_folder, 'audio')

    # 创建输出目录
    os.makedirs(mel_dir, exist_ok=True)
    os.makedirs(linear_dir, exist_ok=True)
    os.makedirs(wav_dir, exist_ok=True)

    # 调用预处理函数生成元数据
    metadata = preprocessor.build_from_path(hparams, input_folders, mel_dir, linear_dir, wav_dir, n_jobs, tqdm=tqdm)

    # 写入 train.txt 文件
    train_txt_path = os.path.join(output_folder, 'train.txt')
    with open(train_txt_path, 'w', encoding='utf-8') as f:
        for m in metadata:
            f.write('|'.join([str(x) for x in m]) + '\n')

    print(f"train.txt 文件已生成，路径：{train_txt_path}")

# 示例调用
if __name__ == "__main__":
    input_folders = ["/content/drive/MyDrive/LJSpeech-1.1"]  # 替换为你的音频数据路径
    output_folder = "/content/drive/MyDrive/training_data"  # 替换为你的输出路径
    n_jobs = 4  # 使用的并行线程数

    generate_train_txt(input_folders, output_folder, n_jobs)
```

```python
import os

def generate_train_txt_from_metadata(output_folder, metadata_path):
    mel_dir = os.path.join(output_folder, 'mels')
    linear_dir = os.path.join(output_folder, 'linear')
    train_txt_path = os.path.join(output_folder, 'train.txt')

    # 检查文件夹是否存在
    if not os.path.exists(mel_dir) or not os.path.exists(linear_dir):
        raise FileNotFoundError("Mel 或 Linear 频谱文件夹不存在，请确认路径是否正确！")

    # 从 metadata.csv 中提取数据并生成 train.txt
    with open(metadata_path, 'r', encoding='utf-8') as meta_file:
        metadata = []
        for line in meta_file:
            parts = line.strip().split('|')
            if len(parts) < 2:
                print(f"跳过不完整的行：{line.strip()}")
                continue

            file_id = parts[0]  # 提取文件 ID
            text = parts[1]  # 提取文本内容

            # 构建 Mel 和 Linear 文件路径
            mel_path = os.path.join(mel_dir, f"mel-{file_id}.npy")
            linear_path = os.path.join(linear_dir, f"linear-{file_id}.npy")

            if os.path.exists(mel_path) and os.path.exists(linear_path):
                # 假设 mel_frames 和 timesteps 为占位值，实际需要根据文件内容计算
                mel_frames = 500  # 这里可以替换为实际计算的值
                timesteps = 1024  # 这里可以替换为实际计算的值
                metadata.append(f"{file_id}|mels/mel-{file_id}.npy|linear/linear-{file_id}.npy|{mel_frames}|{timesteps}|{text}")
            else:
                print(f"跳过 {file_id}，因为对应的 Mel 或 Linear 文件不存在！")

    # 写入 train.txt 文件
    with open(train_txt_path, 'w', encoding='utf-8') as f:
        f.write("\n".join(metadata))

    print(f"train.txt 文件已生成，路径：{train_txt_path}")


# 示例调用
if __name__ == "__main__":
    output_folder = "/content/drive/MyDrive/training_data"  # 替换为存储 mels/ 和 linear/ 的路径
    metadata_path = "/content/drive/MyDrive/LJSpeech-1.1/metadata.csv"  # 替换为 LJSpeech 的 metadata.csv 路径

    generate_train_txt_from_metadata(output_folder, metadata_path)
```
